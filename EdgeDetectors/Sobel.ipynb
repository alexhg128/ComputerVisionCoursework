```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from typing import List
```

```python
def sobel(img:List[List], th = 0.5) -> List[List]:

    w = len(img[0])
    h = len(img)

    gx = [
        [-1, 0, 1],
        [-2, 0, 2],
        [-1, 0, 1]
    ]

    gy = [
        [-1, -2, -1],
        [0, 0, 0],
        [1, 2, 1]
    ]

    stg = np.zeros((h, w))

    m = 0

    for i in range(0, h-2):
        for j in range(0, w-2):
            tmp = img[i:i+3, j:j+3]
            s1 = np.sum(np.matmul(gx, tmp))
            s2 = np.sum(np.matmul(gy, tmp))
            stg[i+1, j+1] = int(np.sqrt((s1 ** 2) + (s2 ** 2)))
            m = m if stg[i+1, j+1] <= m else stg[i+1, j+1]

    m = int(m * th)

    for i in range(h):
        for j in range(w):
            stg[i, j] = 255 if stg[i, j] <= m else 0

    return stg
```

```python
img = cv.imread('../Images/5.jpg', 0)
plt.imshow(img, cmap="gray", vmin=0, vmax=255)
plt.show()
```

```python
edge1 = sobel(img, 0.05)
edge2 = sobel(img, 0.10)
edge3 = sobel(img, 0.15)
edge4 = sobel(img, 0.20)
```

```python
plt.figure()
f, ax = plt.subplots(2,2, figsize=(10,8))

ax[0,0].imshow(edge1, cmap="gray", vmin=0, vmax=255)
ax[0,1].imshow(edge2, cmap="gray", vmin=0, vmax=255)
ax[1,0].imshow(edge3, cmap="gray", vmin=0, vmax=255)
ax[1,1].imshow(edge4, cmap="gray", vmin=0, vmax=255)

ax[0,0].title.set_text('0.05')
ax[0,1].title.set_text('0.10')
ax[1,0].title.set_text('0.15')
ax[1,1].title.set_text('0.20')

plt.show()
```
